```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns

from pandas import DataFrame
```

```python
data = pd.read_csv("eval_data.csv")
data.netfile = data.netfile.apply(lambda x: x.split(os.sep)[1])
data.head()
```

```python
data.columns
```

```python
g = sns.relplot(data=data, kind="line", row="netfile", col="ranked", x="step", y="cum_reward", hue="kind", ci=None)
for axes_row in g.axes:
    for ax in axes_row:
        ax.grid(linestyle="--", which="both")
        for spine in ["top", "right", "bottom", "left"]:
            ax.spines[spine].set_visible(True)
plt.show()
```
