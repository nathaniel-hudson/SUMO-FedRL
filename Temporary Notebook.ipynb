```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("temp_delete_later.csv")
df.head()
```

```python
sns.lineplot(data=df, x="step", y="congestion", hue="tls_id", ci=None)
plt.grid(linestyle="--")
plt.show()
```

```python
sns.lineplot(data=df.query("0 <= step <= 20"), x="step", y="global_rank", hue="tls_id", ci=None)
plt.grid(linestyle="--")
plt.show()
```
