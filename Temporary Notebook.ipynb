```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("temp_delete_later.csv")
df.head()
```

```python
EP = 1
MIN_STEP = 0
MAX_STEP = 50
```

```python
data = df.query(f"{MIN_STEP} <= step <= {MAX_STEP} and episode == {EP}")
data.head()
```

```python
sns.lineplot(data=data, x="step", y="congestion", hue="tls_id", ci=None)
plt.grid(linestyle="--")
plt.title("Congestion")
plt.show()
```

```python
sns.lineplot(data=data, x="step", y="halt_congestion", hue="tls_id", ci=None)
plt.grid(linestyle="--")
plt.title("Halt Congestion")
plt.show()
```

```python
sns.lineplot(data=data, x="step", y="curr_state_mode", hue="tls_id", ci=None)
plt.grid(linestyle="--")
plt.title("Current State (Mode)")
plt.show()
```

```python
sns.lineplot(data=data, x="step", y="curr_state_std", hue="tls_id", ci=None)
plt.grid(linestyle="--")
plt.title("Current State (Standard Deviation)")
plt.show()
```

```python
sns.lineplot(data=data, x="step", y="global_rank", hue="tls_id")
plt.grid(linestyle="--")
plt.show()
```

```python
sns.lineplot(data=data, x="step", y="local_rank", hue="tls_id")
plt.grid(linestyle="--")
plt.show()
```

```python
sns.lineplot(data=data, x="step", y="reward")
plt.grid(linestyle="--")
plt.show()
```
