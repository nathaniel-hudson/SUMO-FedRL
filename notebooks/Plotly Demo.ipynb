```python
import pandas as pd
import numpy as np
import random
demo = {
    "a": np.random.random(size=1000),
    "b": np.random.random(size=1000),
    "c": np.random.random(size=1000),
    "label": []
}
for i in range(1000):
    a, b, c = demo["a"][i], demo["b"][i], demo["c"][i]

    s = a + b + c
    pa, pb, pc = a/s, b/s, 1 - (a/s + b/s)

    label = np.random.choice(["a", "b", "c"], p=[pa, pb, pc])
    demo["label"].append(label)

    # if a > b and a > c:
    #     demo["label"].append("a")
    # elif b > a and b > c:
    #     demo["label"].append("b")
    # else:
    #     demo["label"].append("c")
demo_df = pd.DataFrame.from_dict(demo)
fig2 = px.scatter_ternary(demo_df, a="a", b="b", c="c", color="label")
fig2.show()
```
